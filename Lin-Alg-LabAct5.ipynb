$$ A = \left\{ \begin{array}\\ 4y \\ 3x + 11 y \end{array} \right. $$

```python
import numpy as np
import matplotlib.pyplot as plt

A = np.array([
    [0, 4],
    [3, 11]
])

plt.xlim(-5,15)
plt.ylim(-5,15)

plt.quiver(0,0, A[0,0], A[0,1], angles='xy', scale_units='xy',scale=1, color='blue') 
plt.quiver(0,0, A[1,0], A[1,1], angles='xy', scale_units='xy',scale=1, color='red')


plt.quiver([0,0],[0,0], A[:,0], A[:,1], 
           angles='xy', scale_units='xy',scale=1, 
           color=['blue','cyan'])
plt.show()
```

$$ B = \left\{ \begin{array}\\ 3x +y + 5z \\ 2x + 4y + 6z \\ 3x +6y+ 15z \\ \end{array} \right. $$

```python
B = np.array([
    [3,1,5],
    [2,4,6],
    [8,6,9]
])

fig = plt.figure()

B1 = fig.gca(projection='3d')

B1.set_xlim([-5, 7])
B1.set_ylim([-5, 7])
B1.set_zlim([-5, 7])
origin = (-3,-3,-3)

B1.quiver(origin, origin, origin, B[:,0], B[:,1], B[:,2], arrow_length_ratio=0.05, colors=['red','blue','orange'])

plt.show()
```
